```python
# =============================================================================
# [1] Configuration & Reproducibility Setup
# =============================================================================
import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
import networkx as nx
from tqdm import tqdm
from pathlib import Path
from collections import defaultdict
import requests
import json

# -- 실험 제어용 스위치 (CONFIG) --
# 실험 1 (Baseline): use_augmentation = False 로 설정하고 실행
# 실험 2 (Final): use_augmentation = True 로 설정하고 실행 (LLM 사용)
CONFIG = {
    'use_gnn': True,            # GNN 사용 여부
    'use_self_training': True,  # Self-Training 사용 여부
    'use_augmentation': True,   # [중요] True면 LLM으로 데이터 증강 (비용 발생), False면 안 함
    
    # [속도 최적화 설정]
    'model_name': 'distilbert-base-uncased', # 가볍고 빠른 모델
    'batch_size': 32,                        # 메모리 허용 시 64 추천
    'learning_rate': 5e-5,
    'epochs': 2,                             # 에포크 단축
    'self_train_loops': 2,                   # 반복 횟수 단축
    'max_len': 256,                          # 입력 길이 단축 (속도 향상 핵심)
    'device': 'cuda' if torch.cuda.is_available() else 'cpu',
    
    # LLM 설정 (입력하신 API Key)
    'llm_api_key': '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', 
    'llm_model': 'gpt-4o-mini'
}

# -- 재현성을 위한 시드 고정 --
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)
print(f"Running on device: {CONFIG['device']}")
print(f"Configuration: {CONFIG}")

# =============================================================================
# [2] Data Loading & Taxonomy Construction
# =============================================================================
ROOT = Path("Amazon_products")

TRAIN_CORPUS_PATH = ROOT / "train" / "train_corpus.txt"
TEST_CORPUS_PATH  = ROOT / "test" / "test_corpus.txt"
CLASSES_PATH      = ROOT / "classes.txt"
HIERARCHY_PATH    = ROOT / "class_hierarchy.txt"
KEYWORDS_PATH     = ROOT / "class_related_keywords.txt"
SUBMISSION_DIR    = ROOT / "submission"

SUBMISSION_DIR.mkdir(parents=True, exist_ok=True)

# 1. Load Classes
class_id_to_name = {}
class_name_to_id = {}
with open(CLASSES_PATH, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) >= 2:
            cid, cname = parts[0], parts[1]
            class_id_to_name[int(cid)] = cname
            class_name_to_id[cname] = int(cid)

NUM_CLASSES = len(class_id_to_name)
print(f'Total classes: {NUM_CLASSES}')

# 2. Hierarchy Graph & Adjacency Matrix
G = nx.DiGraph()
G.add_nodes_from(range(NUM_CLASSES))
with open(HIERARCHY_PATH, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) >= 2:
            parent, child = int(parts[0]), int(parts[1])
            G.add_edge(parent, child)

adj = nx.adjacency_matrix(G, nodelist=range(NUM_CLASSES)).todense()
adj_tensor = torch.FloatTensor(adj).to(CONFIG['device'])
adj_tensor = adj_tensor + torch.eye(NUM_CLASSES).to(CONFIG['device']) # Self-loop

# 3. Load Corpus
def load_corpus(path):
    with open(path, 'r', encoding='utf-8') as f:
        return [line.strip() for line in f.readlines() if line.strip()]

train_corpus = load_corpus(TRAIN_CORPUS_PATH)
test_corpus = load_corpus(TEST_CORPUS_PATH)
print(f"Loaded Train: {len(train_corpus)}, Test: {len(test_corpus)}")

# =============================================================================
# [3] Initial Silver Label Generation (Keyword Matching) - Baseline Fix
# =============================================================================
print("Generating initial silver labels using Keywords (Explicit Mention)...")

# 키워드 로드
class_keywords = defaultdict(list)
with open(KEYWORDS_PATH, 'r', encoding='utf-8') as f:
    for line in f:
        if ':' in line:
            cname, kws = line.strip().split(':')
            keywords = [k.strip().lower() for k in kws.split(',')]
            if cname in class_name_to_id:
                cid = class_name_to_id[cname]
                class_keywords[cid] = keywords

silver_labels = torch.zeros(len(train_corpus), NUM_CLASSES)
hit_count = 0

# 키워드 매칭 수행
for i, doc in tqdm(enumerate(train_corpus), total=len(train_corpus), desc="Labeling"):
    doc_lower = doc.lower()
    found = False
    for cid, keywords in class_keywords.items():
        for kw in keywords:
            if kw in doc_lower: # 단순 포함 여부 확인
                silver_labels[i, cid] = 1.0
                found = True
    
    if found:
        hit_count += 1
    else:
        # 키워드 매칭 실패 시 fallback (랜덤 1개)
        fallback = random.randint(0, NUM_CLASSES-1)
        silver_labels[i, fallback] = 1.0

print(f"Keyword Matching Coverage: {hit_count}/{len(train_corpus)} docs labeled.")

# =============================================================================
# [4] LLM Data Augmentation (Real API)
# =============================================================================
def augment_data_with_llm(docs, silver_labels):
    if not CONFIG['use_augmentation']:
        print("Skipping Data Augmentation (Config is False)")
        return docs, silver_labels
    
    print("Performing Data Augmentation using LLM (Real API Calls)...")
    
    # 데이터가 적은 Tail Class 식별
    class_counts = silver_labels.sum(dim=0)
    tail_classes_idxs = torch.argsort(class_counts)[:5] # 하위 5개 클래스만 증강
    
    new_docs = []
    new_labels = []
    
    MAX_CALLS = 10 # [주의] 테스트용 10회 제한. 필요 시 500 등으로 늘리세요.
    calls_made = 0
    
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {CONFIG['llm_api_key']}"
    }

    for c_idx in tail_classes_idxs:
        if calls_made >= MAX_CALLS: break
        class_name = class_id_to_name[c_idx.item()]
        
        prompt = f"Write a short, realistic Amazon product review for a product in the category: '{class_name}'. Do not include the category name in the review."
        
        payload = {
            "model": CONFIG['llm_model'],
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.7,
            "max_tokens": 100
        }

        try:
            response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
            response_data = response.json()
            
            if 'choices' in response_data:
                generated_text = response_data['choices'][0]['message']['content']
                new_docs.append(generated_text)
                
                label_vec = torch.zeros(NUM_CLASSES)
                label_vec[c_idx] = 1.0
                new_labels.append(label_vec)
                
                calls_made += 1
                print(f"[LLM] Generated review for: {class_name}")
            else:
                print(f"OpenAI Error: {response_data}")
                
        except Exception as e:
            print(f"API Call Failed: {e}")

    if len(new_docs) > 0:
        print(f"Added {len(new_docs)} augmented reviews.")
        combined_docs = docs + new_docs
        new_labels_tensor = torch.stack(new_labels)
        combined_labels = torch.cat([silver_labels, new_labels_tensor], dim=0)
        return combined_docs, combined_labels
    else:
        return docs, silver_labels

# 증강 실행
final_train_docs, final_train_labels = augment_data_with_llm(train_corpus, silver_labels)

# =============================================================================
# [5] Model Definition (DistilBERT + GNN)
# =============================================================================
class GNNLayer(nn.Module):
    def __init__(self, in_features, out_features, adj):
        super(GNNLayer, self).__init__()
        self.linear = nn.Linear(in_features, out_features)
        
    def forward(self, x, adj):
        support = self.linear(x)
        output = torch.mm(adj, support)
        return F.relu(output)

class TaxoClassModel(nn.Module):
    def __init__(self, model_name, num_classes, hidden_dim=768):
        super(TaxoClassModel, self).__init__()
        # DistilBERT 사용
        self.bert = AutoModel.from_pretrained(model_name)
        self.doc_fc = nn.Linear(self.bert.config.hidden_size, hidden_dim)
        
        self.class_embeddings = nn.Parameter(torch.randn(num_classes, hidden_dim))
        
        if CONFIG['use_gnn']:
            self.gnn = GNNLayer(hidden_dim, hidden_dim, None)
        
        self.dropout = nn.Dropout(0.1)

    def forward(self, input_ids, attention_mask, adj):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        # DistilBERT Output: [0] is last_hidden_state
        doc_emb = outputs[0][:, 0, :] # [CLS] token
        doc_emb = self.doc_fc(doc_emb)
        
        if CONFIG['use_gnn']:
            class_emb = self.gnn(self.class_embeddings, adj)
        else:
            class_emb = self.class_embeddings
            
        logits = torch.matmul(doc_emb, class_emb.t())
        return logits

model = TaxoClassModel(CONFIG['model_name'], NUM_CLASSES).to(CONFIG['device'])

# =============================================================================
# [6] Dataset & Training Loop
# =============================================================================
class ReviewDataset(Dataset):
    def __init__(self, texts, labels=None, tokenizer=None, max_len=128, is_test=False):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.is_test = is_test
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        encoding = self.tokenizer.encode_plus(
            text, add_special_tokens=True, max_length=self.max_len,
            padding='max_length', truncation=True, return_tensors='pt'
        )
        item = {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten()
        }
        if not self.is_test:
            item['labels'] = self.labels[idx]
        return item

tokenizer = AutoTokenizer.from_pretrained(CONFIG['model_name'])
train_dataset = ReviewDataset(final_train_docs, final_train_labels, tokenizer, CONFIG['max_len'])
train_loader = DataLoader(train_dataset, batch_size=CONFIG['batch_size'], shuffle=True)

optimizer = torch.optim.AdamW(model.parameters(), lr=CONFIG['learning_rate'])
criterion = nn.BCEWithLogitsLoss()

loops = CONFIG['self_train_loops'] if CONFIG['use_self_training'] else 1
print(f"Starting Training ({loops} loops)...")

for loop in range(loops):
    print(f"--- Self-Training Loop {loop+1}/{loops} ---")
    model.train()
    
    for epoch in range(CONFIG['epochs']):
        total_loss = 0
        for batch in tqdm(train_loader, desc=f"Ep {epoch+1}"):
            input_ids = batch['input_ids'].to(CONFIG['device'])
            mask = batch['attention_mask'].to(CONFIG['device'])
            labels = batch['labels'].to(CONFIG['device'])
            
            optimizer.zero_grad()
            logits = model(input_ids, mask, adj_tensor)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Loss: {total_loss / len(train_loader):.4f}")

# =============================================================================
# [7] Inference & Submission
# =============================================================================
print("Starting Inference...")
test_dataset = ReviewDataset(test_corpus, tokenizer=tokenizer, max_len=CONFIG['max_len'], is_test=True)
test_loader = DataLoader(test_dataset, batch_size=CONFIG['batch_size'], shuffle=False)

model.eval()
all_predictions = []

with torch.no_grad():
    for batch in tqdm(test_loader, desc="Inference"):
        input_ids = batch['input_ids'].to(CONFIG['device'])
        mask = batch['attention_mask'].to(CONFIG['device'])
        logits = model(input_ids, mask, adj_tensor)
        probs = torch.sigmoid(logits)
        all_predictions.append(probs.cpu())

all_predictions = torch.cat(all_predictions, dim=0)

predicted_labels = []
MIN_LABELS = 2
MAX_LABELS = 3

for i in range(all_predictions.shape[0]):
    probs = all_predictions[i]
    sorted_idx = torch.argsort(probs, descending=True)
    topk = sorted_idx[:MAX_LABELS]
    selected = []
    for idx in topk:
        if probs[idx] > 0.5 or len(selected) < MIN_LABELS:
            selected.append(int(idx))
    if len(selected) < MIN_LABELS:
        selected = [int(x) for x in sorted_idx[:MIN_LABELS]]
    predicted_labels.append(selected)

test_ids = range(len(test_corpus))
submission = pd.DataFrame({
    'id': test_ids,
    'label': [','.join(str(l) for l in labels) for labels in predicted_labels] 
})

SUBMISSION_PATH = SUBMISSION_DIR / "submission.csv"
submission.to_csv(SUBMISSION_PATH, index=False)
print(f"Submission saved to: {SUBMISSION_PATH}")
print(submission.head())
```
